```python
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt
```

```python
SC = pd.read_csv('ResultsMediumClassicS.csv')
QC = pd.read_csv('ResultsMediumClassicQ.csv')
SAPPROXC = pd.read_csv('ResultsMediumClassicSAPPROX.csv')
QAPPROXC = pd.read_csv('ResultsMediumClassicQAPPROX.csv')
SAPPROXC2 = pd.read_csv('ResultsMediumClassicSAPPROX2.csv')
QAPPROXC2 = pd.read_csv('ResultsMediumClassicQAPPROX3.csv')
APPROXC= pd.read_csv('ResultsMediumClassicAPPROX.csv')

SM = pd.read_csv('ResultsMediumGridSARSA.csv')
QM = pd.read_csv('ResultsMediumGridQ.csv')
SAPPROXM = pd.read_csv('ResultsMediumGridSARSAPROX.csv')
QAPPROXM = pd.read_csv('ResultsMediumGridQPROX.csv')
APPROXM= pd.read_csv('ResultsmediumGridAPPROX.csv')

SS = pd.read_csv('ResultssmallGridS.csv')
QS = pd.read_csv('ResultssmallGridQ.csv')
SAPPROXS = pd.read_csv('ResultssmallGridSAPPROX.csv')
QAPPROXS = pd.read_csv('ResultssmallGridQAPPROX.csv')
APPROXS= pd.read_csv('ResultssmallGridAPPROX.csv')
```

```python
x = 30

groupSC = SC.groupby(np.arange(len(SC))//100).mean()
groupQC = QC.groupby(np.arange(len(QC))//100).mean()
groupSAPPROXC = SAPPROXC.groupby(np.arange(len(SAPPROXC))//x).mean()
groupQAPPROXC = QAPPROXC.groupby(np.arange(len(QAPPROXC))//x).mean()
groupSAPPROXC2 = SAPPROXC2.groupby(np.arange(len(SAPPROXC2))//x).mean()
groupQAPPROXC2 = QAPPROXC2.groupby(np.arange(len(QAPPROXC2))//x).mean()
groupAPPROXC = APPROXC.groupby(np.arange(len(APPROXC))//x).mean()

groupSM = SM.groupby(np.arange(len(SM))//100).mean()
groupQM = QM.groupby(np.arange(len(QM))//100).mean()
groupSAPPROXM = SAPPROXM.groupby(np.arange(len(SAPPROXM))//100).mean()
groupQAPPROXM = QAPPROXM.groupby(np.arange(len(QAPPROXM))//100).mean()
groupAPPROXM = APPROXM.groupby(np.arange(len(APPROXM))//x).mean()

groupSS = SS.groupby(np.arange(len(SS))//100).mean()
groupQS = QS.groupby(np.arange(len(QS))//100).mean()
groupSAPPROXS = SAPPROXS.groupby(np.arange(len(SAPPROXS))//100).mean()
groupQAPPROXS = QAPPROXS.groupby(np.arange(len(QAPPROXS))//100).mean()
groupAPPROXS = APPROXS.groupby(np.arange(len(APPROXS))//x).mean()
```

```python
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(10)
plt.title("Score sur smallGrid")
plt.plot(range(len(SS)//100) , groupSS['Score'], label = 'SARSA')
plt.plot(range(len(QS)//100) , groupQS['Score'], label = 'Q-Learning')
plt.plot(range(len(SAPPROXS)//100) , groupSAPPROXS['Score'], label = 'Approximate SARSA')
plt.plot(range(len(QAPPROXS)//100) , groupQAPPROXS['Score'], label = 'Approximate Q-Learning')
plt.plot(range(len(groupAPPROXS)) , groupAPPROXS['Score'], label = 'True Approximate Q-Learning')

plt.xlabel("parties (10^2)")
plt.ylabel("Moyenne de Score sur 100 parties")
plt.legend()
plt.savefig('Score_sur_smallGrid')
plt.show()
```

```python
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(10)
plt.title("WinRate sur smallGrid")
plt.plot(range(len(SS)//100) , groupSS['WOR'], label = 'SARSA')
plt.plot(range(len(QS)//100) , groupQS['WOR'], label = 'Q-Learning')
plt.plot(range(len(SAPPROXS)//100) , groupSAPPROXS['WOR'], label = 'Approximate SARSA')
plt.plot(range(len(QAPPROXS)//100) , groupQAPPROXS['WOR'], label = 'Approximate Q-Learning')
plt.plot(range(len(groupAPPROXS)) , groupAPPROXS['WOR'], label = 'True Approximate Q-Learning')


plt.plot(range(len(SS)//100), [0.8 for i in range(len(SS)//100)], label = '80% WR')

plt.xlabel("parties (10^2)")
plt.ylabel("Moyenne de WinRate sur 100 parties")
plt.legend()
plt.savefig('WR_sur_smallGrid')
plt.show()

print("WR sur 1000 parties aprs entranement:")
#meanSARSA = test["WOR"].iloc[-10:]

print("SARSA: "+ str(groupSS["WOR"].iloc[-1:]))
print("Q-Learning: "+ str(groupQS["WOR"].iloc[-1]))
print("Approximate SARSA: "+ str(groupSAPPROXS["WOR"].iloc[-1]))
print("Approximate Q-Learning: "+ str(groupQAPPROXS["WOR"].iloc[-1]))
print("True Approximate Q-Learning: "+ str(groupAPPROXS["WOR"].iloc[-1]))
```
